```python
import urllib.request
#접근하려는 사이트 주소 객체 생성
nateUrl = 'https://www.nate.com//'
#해당 웹사이트의 모든 html 태그 읽기
htmlObject = urllib.request.urlopen(nateUrl)
#웹사이트 모든 html 태그를 읽기
html = htmlObject.read()
#출력
print(html)
```

```python
import bs4
import urllib.request
#접근하려는 사이트 주소 객체 생성
nateUrl = 'https://www.nate.com//'
#해당 웹사이트의 모든 html 태그 읽기
htmlObject = urllib.request.urlopen(nateUrl)
#정리된 태그들의 트리
bsObject=bs4.BeautifulSoup(htmlObject, 'html.parser')
print(bsObject)
```
